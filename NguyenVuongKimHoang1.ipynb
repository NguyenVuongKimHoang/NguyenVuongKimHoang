```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(6)
y=x+1
plt.plot(x,y,'o--')
plt.plot(x,-y,'o-')
plt.title('y=x va y=-x')
plt.grid('x')
plt.xlabel('x')
plt.ylabel('y')
plt.show
```
