```python

import matplotlib.pyplot as plt
import numpy as np
x=np.arange(7)
plt.plot(x,-x**2,x,-x**3,x,-2*x,x,x-2**x)
plt.grid(True)
plt.savefig('test.png')
plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(3)
plt.subplots_adjust(wspace=0.3,hspace=0.3)
plt.subplot(3,2,1)
plt.plot(x,x**2,color='g')
plt.subplot(2,2,2)
plt.plot(x,x**3,color='g')
plt.subplot(2,2,3)
plt.plot(x,-2*x,color='b')
plt.subplot(2,2,4)
plt.plot(x,-2**x,'o')
plt.plot()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.axis([0,100,0,10])
plt.scatter(60,5,1000,color='b',alpha=1)
plt.scatter(80,5,s=1000,color='r',alpha=.5)
plt.scatter(100,5,s=1000,color='g',alpha=.1)
plt.show()
```
